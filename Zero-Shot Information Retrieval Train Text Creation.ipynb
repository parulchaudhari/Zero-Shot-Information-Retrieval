```python
import os
import pandas as pd

df = pd.DataFrame(columns=['Text', "FileName",'Folder'])
file_not_found = 0
path = r"C:\Users\pchaudhari\OneDrive - VMware, Inc\VMwareCorp\Documents\MLPO\IR\2021-input"
```

```python
import pathlib
import glob

pd.set_option('display.max_columns', None)
pd.set_option("max_rows", None)

train_files = pathlib.Path(r"C:\Users\pchaudhari\OneDrive - VMware, Inc\VMwareCorp\Documents\MLPO\IR\2021-input")
```

```python
files_count_train = len(list(train_files.glob('*/*.txt')))
files_count_train
```

```python
for filepath in train_files.glob('*/*.txt'):
        try:
            with open(filepath) as f:
                observation = f.read()
                current_df = pd.DataFrame({'Text': [observation],"FileName":[os.path.basename(filepath)],'Folder':[os.path.basename(os.path.dirname(filepath))]})
                df = df.append(current_df, ignore_index=True)
        except:
            file_not_found= file_not_found+1
            print(filepath)
            #break
```
